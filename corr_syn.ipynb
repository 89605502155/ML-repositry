```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.DataFrame(file1['B0'],columns=['first','second','third','fourth'])
df1
```

```python
df1.corr()
```

```python
corr = df1.corr()
corr.style.set_precision(2).background_gradient(cmap='coolwarm')
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
sns.heatmap(corr,annot=True, fmt='.4f', 
            cmap=plt.get_cmap('coolwarm'), cbar=False, ax=ax,
            annot_kws={'size':30})
ax.set_yticklabels(ax.get_yticklabels(), rotation="horizontal",fontsize=30)
ax.set_xticklabels(ax.get_xticklabels(), rotation="horizontal",fontsize=30)
ax.get_xaxis().set_tick_params(direction='in')
ax.get_yaxis().set_tick_params(direction='in')
ax.set_ylabel(' ',labelpad=100)
# ax.title("Emission coef corr",fontsize=30)
ax.set_title("Emission coef corr", fontsize=28,loc="center" ,pad=15)
# ax.tick_params(labelsize=20)

plt.savefig('syn_corr_emission_four.svg', format='svg',pad_inches=50)
plt.savefig('syn_corr_emission_four.pdf', format='pdf',dpi=300)
# ax.set_xlabel(ax.get_xlabel,fontsize=25)
```
